```python
%matplotlib inline

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pathlib import Path
```

# Do we think the same thing is happening in all of these countries?

This is a hypothetical dataset representing pesticide residues above a safety threshold vs. percentage of organic farms in different countries. How can we tell if trends are consistent across countries?

```python
for file in Path('.').glob("*.csv"):
    df = pd.read_csv(file)
    x, y = df['x'], df['y']
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    coeff = np.polyfit(x, y, deg=1)
    corr = np.corrcoef(x, y)[0, 1]

    print(f"For {file}: x_mean={x_mean:.3f}, y_mean={y_mean:.3f}, linear regression fr(x)={coeff[1]:.1f} {coeff[0]:.2f} x, r = {corr:.3f} ")
```

```python
for file in Path('.').glob("*.csv"):
    df = pd.read_csv(file)
    x, y = df['x'], df['y']
    plt.scatter(df['x'], df['y'], marker='o')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
x_mean_values = []
y_mean_values = []

for file in Path('.').glob("*.csv"):
    df = pd.read_csv(file)
    x, y = df['x'], df['y']
    xm = np.mean(x)
    ym = np.mean(y)
    x_mean_values.append(xm)
    y_mean_values.append(ym)

df = pd.DataFrame(zip(x_mean_values, y_mean_values), columns = ("x_mean", "y_mean"))

plt.scatter(df['x_mean'], df['y_mean'], marker='o')
plt.xlabel("x-mean")
plt.ylabel("y-mean")
plt.show()
```
