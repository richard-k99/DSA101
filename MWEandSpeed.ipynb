`````python
# Now we turn the definition into Python code.

def fib(n):
    if n == 0:
        return 0
    elif n < 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
`````

`````python
[fib(n) for n in range(10)]
`````

`````python
# Let's see how long this takes...

import time

for n in range(30):
    start = time.time()
    result = fib(n)
    elapsed = time.time() - start
    print(f"fib({n:2d}) = {result:8d}, elapsed time: {elapsed:.3f}s")
`````

Let it run a bit further...

Now we add one line for speed (ok, two). :)

`````python
# Let's speed it up!

from functools import lru_cache

@lru_cache
def fib(n):
    if n == 0:
        return 0
    elif n < 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)

for n in range(50):
    start = time.time()
    result = fib(n)
    elapsed = time.time() - start
    print(f"fib({n:2d}) = {result:8d}, elapsed time: {elapsed:.3f}s")
`````
